```python
# Find duplicates based on the Email column
duplicates = df_graduation[df_graduation.duplicated(subset='Email', keep=False)]

# Display the count of duplicates
print("Count of duplicate rows based on Email column:", duplicates.shape[0])

# Drop the duplicates, keeping the first occurrence
df_graduation = df_graduation.drop_duplicates(subset='Email', keep='first')

# Display the number of rows after dropping duplicates
print("Number of rows after dropping duplicates:", df_graduation.shape[0])
```

```python
# Convert 'Created' column to datetime
df_graduation['Created'] = pd.to_datetime(df_graduation['Created'], format='%m/%d/%Y %I:%M:%S %p')

# Extract the year and create a new column 'Created Year'
df_graduation['Created Year'] = df_graduation['Created'].dt.year

df_graduation.head()
```

```python
df_graduation['Branch Names'] = df_graduation['Other Branch'].combine_first(df_graduation['Branch/ Specialisation'])
df_graduation['Branch Names'].unique()
```

```python
df_graduation.isna().sum()
```

```python
df_graduation['Academic Year'].unique()
```

```python
# Renaming the column
df_graduation.rename(columns={'What is your current academic year?': 'Current Academic Year'}, inplace=True)
df_graduation['Current Academic Year'].unique()
```

```python
df_graduation['Current Academic Year'] = df_graduation['Current Academic Year'].replace({
    'Final Year': 4,
    '1st Year': 1,
    '2nd Year': 2,
    '3rd Year': 3
})

# filled nan values with '1'
df_graduation['Current Academic Year'].fillna(1, inplace=True)
df_graduation['Current Academic Year'].unique()
```

```python
# Drop rows with NaN values in 'Academic Year' column
df_graduation = df_graduation.dropna(subset=['Academic Year'])
```

```python
df_graduation.head()
```

```python
df_graduation.shape
```

```python
df_graduation['Graduation Year'] = (df_graduation['Created Year']- df_graduation['Current Academic Year']) + df_graduation['Academic Year'] 

df_graduation['Graduation Year'] = df_graduation['Graduation Year'].astype(int)

df_graduation.head()
```

```python
def calGraduationYear(createdYear, academicYear, currentAY):
    graduationY = (createdYear - currentAY) + academicYear
    
    return graduationY
```
